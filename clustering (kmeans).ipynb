```python
#импорт библиотек
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_blobs

#генерируем точки для матрицы D случайным образом
np.random.seed(100)
X1 = np.random.normal(loc=[0,-10], size=(100,2)) 
X2 = np.random.normal(loc=[-10, 0], size=(100,2)) 
X3 = np.random.normal(loc=[0, 0], size=(100,2)) 
X = np.vstack((X1,X2,X3)) 
y = np.array([1]*100 + [2]*100 + [3]*100) 
#задаем k кластер 
k_means = KMeans(n_clusters=3) 

clusters = k_means.fit_predict(X) 

pred = np.zeros(X.shape[0])

plt.scatter(X[:, 0], X[:, 1], c=clusters) 
plt.title('K-Means') 
plt.show()

fig = plt.figure()
ax = fig.add_subplot(111)
#plt.figure(figsize=(12, 12)) #задали размер области для вывода кластеров 
for t in range(1,10): #организуем повторения 10 раз
 k_means = KMeans(n_clusters=t) 

 clusters = k_means.fit_predict(X) 

plt.subplot(3, 2, t-1) 
plt.scatter(X[:, 0], X[:, 1], c=clusters) 
plt.title('n_clusters = {}'.format(t)) 
plt.show()
```
